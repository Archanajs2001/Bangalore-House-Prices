```python
#replacing 'bedroom' with 'BHK'

import re

def convert_bedroom_to_bhk(text):
    # Use regular expression to replace 'bedroom' with 'bhk'
    if isinstance(text, str):
      result = re.sub(r'(\d*)\s*Bedroom', r'\1 BHK', text)
      return result
    else:
        # If not a string, return the original value
        return text
```

```python
data['size'] = data['size'].apply(convert_bedroom_to_bhk)
```

```python
#function to check whether object is float
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data[~data['total_sqft'].apply(is_float)]
```

```python
#to convert continous value to discrete
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convert_sqft_to_num)
data = data[data.total_sqft.notnull()]
data.head()
```

```python
data.isnull().sum()
```

```python
#removing null values from columns that are important
data = data[data['size'].notnull()]
data = data[data['location'].notnull()]
data = data[data['bath'].notnull()]
```

```python
data['size'].value_counts()
```

```python
#making separate dataframes for top 4 sizes
one_bhk = pd.DataFrame(data[data['size' ] == '1 BHK'])
two_bhk = pd.DataFrame(data[data['size' ] == '2 BHK'])
three_bhk = pd.DataFrame(data[data['size' ] == '3 BHK'])
four_bhk = pd.DataFrame(data[data['size' ] == '4 BHK'])
```

```python
#converting datasets to csv file
one_bhk.to_csv('1_bhk.csv', index = False)
two_bhk.to_csv('2_bhk.csv', index = False)
three_bhk.to_csv('3_bhk.csv', index = False)
four_bhk.to_csv('4_bhk.csv', index = False)
```
